```python
# Transforming the data non-numerical intonumerical
from sklearn.preprocessing import LabelEncoder
employee_turnover = pd.DataFrame(employee_turnover)

for column in employee_turnover.columns:
    if employee_turnover[column].dtype == np.number:
        continue
    employee_turnover[column] = LabelEncoder().fit_transform(employee_turnover[column])    
```

```python
# Creating a new column
employee_turnover['Age_Year']= employee_turnover['Age']
```

```python
#removing age column
employee_turnover = employee_turnover.drop('Age', axis=1)
```

```python
# Splitting the data into independent 'x' and dependent 'y'
x= employee_turnover.iloc[:, 1:employee_turnover.shape[0]].values
y= employee_turnover.iloc[:, 0].values
```

```python
# Splitting the training and testing set 80 and 20 respectively
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=0)
```

```python
print(x.shape, x_train.shape, x_test.shape)
```

# Data modelling

```python
# estimating random forest 
from sklearn.ensemble import RandomForestClassifier
forest= RandomForestClassifier(n_estimators=10, criterion='entropy', random_state=0)
forest.fit(x_train, y_train)
```

```python
# accuracy of the training data
forest.score(x_train, y_train)
```

```python
# confusion matrix and accuracy of testing data
from sklearn.metrics import confusion_matrix
cm= confusion_matrix(y_test, forest.predict(x_test))
TN=cm[0, 0]
TP=cm[1, 1]
FN=cm[1, 0]
FP=cm[0, 1]
print(cm)
accuracy=(TN+TP)/(TN+TP+FN+FP)
print('accuracy of testing data:{:.2f}'.format(accuracy*100))
```
